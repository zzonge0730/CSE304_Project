```python
import numpy as np
import matplotlib.pyplot as plt

Z = np.load("features/Z_pca.npy")
y = np.load("features/y.npy")

plt.figure(figsize=(6,6))
plt.scatter(Z[:,0], Z[:,1], c=y, cmap='coolwarm', s=60, edgecolor='k')
plt.title("PCA Compressed Features (Labelled)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
